## Task 2:
Find the number of user reviews and the average rating for each book.

```python

query = """

SELECT 
  b.title,
  COUNT(DISTINCT rev.review_id) AS num_reviews, 
  ROUND(AVG(rat.rating)::numeric, 2) AS avg_rating
FROM 
  books AS b
  LEFT JOIN ratings AS rat USING(book_id)
  LEFT JOIN reviews AS rev USING(book_id)
GROUP BY 
  b.title
ORDER BY 
  num_reviews DESC;
"""
df = pd.read_sql(query, con=engine)

# print the DataFrame
df
```

the book with the most reviews is Memoirs of a Geisha with 8 reviews and a 4.14 average rating. there are also books that don't have reviews but are rated. Let's look at them separately

## Task 3
Identify the publisher that has released the greatest number of books with more than 50 pages (this will help you exclude brochures and similar publications from your analysis).

```python

query = """
SELECT 
  publishers.publisher,
  COUNT(books.book_id) AS num_books
FROM 
  books
  INNER JOIN publishers ON books.publisher_id = publishers.publisher_id
WHERE 
  books.num_pages > 50
GROUP BY 
  publishers.publisher
ORDER BY 
  num_books DESC;
"""
df = pd.read_sql(query, con=engine)

# print the DataFrame

df

```

 We see that we have 334 such editions.  The leader is Penguin Books with 42 published books over 50 pages.

## Task 4:
Identify the author with the highest average book rating: look only at books with at least 50 ratings.

```python


query = """
SELECT 
  authors.author, 
  ROUND(AVG(ratings.rating), 2) AS avg_rating
FROM 
  ratings
  INNER JOIN books ON ratings.book_id = books.book_id
  INNER JOIN authors ON books.author_id = authors.author_id
WHERE 
  ratings.book_id IN (
    SELECT 
      book_id 
    FROM 
      ratings 
    GROUP BY 
      book_id 
    HAVING 
      COUNT(*) >= 50
  )
GROUP BY 
  authors.author
ORDER BY 
  avg_rating DESC
LIMIT 5;
"""
df = pd.read_sql(query, con=engine)

# print the DataFrame

df
```

## Task 5:
Find the average number of text reviews among users who rated more than 50 books.

```python
           
 
query = """
SELECT
  ROUND(AVG(Sub_rev.reviews_avg)::numeric, 2) AS users_avg_reviews
FROM
  (SELECT
    rev.username AS username,
    COUNT(rev.review_id) AS reviews_avg
  FROM
    reviews rev
  WHERE
    rev.username IN
      (SELECT
        Sub_rat.username
      FROM
        (SELECT
          rat.username AS username,
          COUNT(rat.rating_id) AS ratings_num
        FROM
          ratings rat
        GROUP BY
          rat.username
        HAVING
          COUNT(rat.rating_id) > 50) AS Sub_rat)
  GROUP BY
    rev.username) AS Sub_rev;
"""
df = pd.read_sql(query, con=engine)

# print the DataFrame

df           
```
